Matlab style interface

```python

f = lambda x: np.sin(x)

# the number 10 (in this case) decides the pointiness of the curve, if we would of hade 1000 it would be more exact 
x = np.linspace(-5,5, 10)
plt.figure(dpi = 200)
# '--o' points out the 10 points we have asked for earlier 
plt.plot(x, f(x),'--o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("$\\sin{(x)}$")
# if you type sin x this way you come in to latesh mode witch makes the x cursive, you can also do to the power of 2 (x^2)
```

oop approach 

```python
fig, ax = plt.subplots(1)

x = np.linspace(-5,5, 100)
ax.plot(x, f(x))
ax.set(xlabel = "x", ylabel="y", title="$\\sin{x}$")
```

## subplots


```python
fig,ax = plt.subplots(1,2, figsize = (16,4), dpi = 200)
ax[0].plot(x, np.sin(x), label = "sin(x)")
ax[0].plot(x, np.cos(x), label = "cos(x)")
ax[0].legend()
ax[0].set(title = "sin(x) and cos(x)", xlabel = "x", ylabel = "y")


ax[1].plot(x, np.cos(x))
ax[0].set(title = "cos(x)")
```

## Histogram and bubble chart

```python
fig, ax = plt.subplots(1, 2, dpi=150, figsize=(10, 4))

rng = np.random.RandomState(42)
# randn - draws random number from normal distrubution
data_normal_dist = rng.randn(1000)

ax[0].hist(data_normal_dist, bins=30, alpha=0.5)
ax[0].set(title="Histogram", xlabel="x", ylabel="y")

size = 1000 * rng.rand(100)
sc = ax[1].scatter(
    x=rng.randn(100),
    y=rng.randn(100),
    c=rng.randn(100),
    alpha=0.3,
    cmap="viridis",
    s=size,
)

fig.colorbar(sc)
ax[1].set(title="Bubble chart", xlabel="x", ylabel="y")

```
