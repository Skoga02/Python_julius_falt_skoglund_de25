```python
from rectangle import Rectangle 
from circle import Circle

r1= Rectangle(4, 5, x=2, y=3)
r2= Rectangle(5, 5)
c1= Circle(3)
c2= Circle(1)

print(r1, "Area:", r1.area)
print(c1, "perimeter:", c1.perimeter)
print("r2 is square", r2.is_square())
print("c2 is unit circle?", c2.is_unit_circle())

print("Comprisons:")
print("r1 == r2:", r1 == r2)
print("c1 > c2:", c1 > c2)

r1.draw()
c1.draw()
```
